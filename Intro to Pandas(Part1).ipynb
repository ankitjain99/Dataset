# Reading Data from table 

```python
# conventional way to import pandas
import pandas as pd
# conventional way to import numpy
import numpy as np
```

```python
# read a dataset of movies directly from a URL and store the results in a DataFrame
movies=pd.read_table('http://bit.ly/moviebyuser')
```

```python
# examine the first 5 rows
movies.head()
```

```python
# read a dataset of movie reviewers (modifying the default parameter values for read_table like column names,sepeartor)
user_cols=['S. No.','Age','Gender','Work','Pin']
movies=pd.read_csv('http://bit.ly/moviebyuser',sep='|',header=None,names=user_cols,skiprows=[3,4])
```

```python
movies
```

```python
# select the 'Age' Series using bracket notation
movies['Age']

# or equivalently, use dot notation
movies.Age
```

```python
# create a new 'gAge' Series by concatenation of columns
movies['gAge']=movies['Work'] +','+ movies['Gender']
movies
```
