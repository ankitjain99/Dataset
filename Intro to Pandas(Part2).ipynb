# Removing columns from Data

```python
# conventional way to import pandas
import pandas as pd
```

```python
# read a dataset of top-rated tweets into a DataFrame
df=pd.read_csv('http://bit.ly/tweetsdf')
```

```python
# show the first 5 rows
df.head()
```

```python
# shows the columns of a DataFrame
df.columns
```

```python
# sort the values of a column in the dataframe, with arguments
df.number_of_shares.sort_values(ascending=False)
# or equivalent use bracket notation
df[number_of_shares].sort_values(ascending=False)
```

```python
# remove a single column (axis=1 refers to columns)
df.drop('content',axis=1)
```

```python
# remove multiple columns at once(using lists)
df.drop(['author', 'number_of_shares'], axis=1)
```
