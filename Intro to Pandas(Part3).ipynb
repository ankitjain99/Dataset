```
# conventional way to import pandas
import pandas as pd
```

```
movie_cols = ['S. No.', 'Age', 'Gender', 'Work', 'Pin']
movie=pd.read_csv('http://bit.ly/moviebyuser',sep='|',header=None,names=movie_cols)
movie.head()   # Example of a method in Pandas
```

```
# Example of column attribute/function in Pandas
movie.columns
```

```
# create a list in which each element refers to a DataFrame row: True if the row satisfies the condition, False otherwise
booleans=[]
for length in movie.Age:
    if length>=25:
        booleans.append(True)
    else:
        booleans.append(False)
```

```
booleans
```

```
# simplify the steps above: no need to write a for loop to create 'is_long' since pandas will broadcast the comparison
is_age = movie.Age >= 25
movie[is_age]

# or equivalently, write it in one line (no need to create the 'is_long' object)
movie[movie.Age >= 25]
```

```
# Use the 'isin' method
movie[movie.Work.isin(['student', 'writer'])]
```
