# Remaining string functions


```python
# conventional way to import pandas
import pandas as pd
```

```python

tweet=pd.read_csv('http://bit.ly/tweetsdf')
tweet.head()
```

```python
# string function replace, replacing 'life' with 'wife'. The first string is with and second string is whom
tweet.content.str.replace('Life','Wife')
```

```python
# groupby function to group functions by number_of_shares
tweet.groupby('author').number_of_shares.max()
```

```python
# count of the values of tweets by author
tweet.author.value_counts()
```

```python
# changing the index to author
tweet.set_index('author')
```
